## Market Basket Analysis

### -- Prepare Data
Import Package

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Load Data
<br>
*Data Source: https://www.kaggle.com/sulmansarwar/transactions-from-a-bakery*

```python
df = pd.read_csv("data/transactions_from_a_bakery.csv")
df.head(5)
```

### -- Clean Data

```python
df.dtypes # Check data type
```

```python
df['Item'] = df['Item'].str.strip() # Remove the spaces at the beginning and the end
df['Item'] = df['Item'].str.title() # Capitalize the first letter of every word
df['Transaction'] = df['Transaction'].astype(str) # Convert integer to string
```

### -- Visualize Data
Top 10 Best Sellers

```python
top_item = df['Item'].value_counts()[:10] # Count item sales and get the top 10
top_item.plot(kind='bar',figsize=(9,6)) # Create bar chart
plt.title('Top 10 Best Sellers') # Add plot title
```

### -- Basket Analysis
List all items in the column, write an order transaction per row. <br>
Return **1** if the item was bought; otherwise, return **0**.

```python
# basket_df = pd.DataFrame()
basket_df = (df
                    .groupby(['Transaction','Item'])['Transaction']
                    .nunique().unstack().reset_index().fillna(0)
                    .set_index('Transaction'))
basket_df.head()
```

```python
basket_df_1 = basket_df.astype(int)
basket_df_1 = basket_df_1.reset_index().rename_axis(None, axis=1)
basket_df_1.head(2)
```

Transform binary variables into text, and display purchased baskets in a **Basket** column.

```python
item_df = df.groupby('Item').Transaction.nunique().reset_index() # Count item sales

col_item1 = basket_df_1.columns.get_loc(item_df.iloc[0,0]) # For loop purposes, get the location of the first item's column

# Set 0 to blank, and 1 to the correlated column name and "+" at the end for later concatenation
for col in basket_df_1.columns[col_item1:]:
    basket_df_1.loc[basket_df_1[col]==0, col] = ''
    basket_df_1.loc[basket_df_1[col]==1, col] = col + ' + '

basket_df_1['Basket']="" # Create a blank Basket column
col = basket_df_1.columns.get_loc('Basket') # For loop purposes, get the location of the Basket column

# Concatenate all the items in the Basket column
r=0
while r < basket_df_1.shape[0]:
    txt = ""
    c=col_item1
    while c < basket_df_1.shape[1]:
        txt = txt + basket_df_1.iloc[r,c]
        c += 1
    basket_df_1.iloc[r,col] = txt
    r += 1

basket_df_1['Basket'] = basket_df_1['Basket'].str[:-3] # Take out the "+" at the end of the basket string

basket_df_1.head(2)
```

Dataframe of Basket data

```python
result = basket_df_1.groupby(by='Basket', as_index=False).agg({'Transaction': pd.Series.nunique}) # Count the number of transactions per basket
result.columns = ['Basket','# of Transactions'] # Rename the dataframe
result['% of Total'] = result['# of Transactions'] / result['# of Transactions'].sum() # Calculate the percentage of each basket in the total
result = result.sort_values(by=['# of Transactions'], ascending = False).reset_index(drop=True) # Sort Result table descending by number of transactions
result.head(10)
```

Visualize the top 10 most popular baskets

```python
plt.figure(figsize=(9, 6)) # Set chart size
ax = sns.barplot(x="# of Transactions", y="Basket", data=result.head(10), palette="Spectral") # Use seaborn package create bar chart

# Formate chart
plt.title('Top 10 Basket')
plt.xlabel('# of Transactions')
plt.ylabel('Basket')
```
